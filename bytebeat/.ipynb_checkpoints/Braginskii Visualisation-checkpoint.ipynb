```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = X
V = Y

dudx = np.gradient(U, axis=1)
dudy = np.gradient(U, axis=0)
dvdx = np.gradient(V, axis=1)
dvdy = np.gradient(V, axis=0)

W11 = 2*dudx
W12 = dudy + dvdx
W21 = W12
W22 = 2*dvdy

theta = 0.4
bx = np.cos(2.0*np.pi*theta)
by = np.sin(2.0*np.pi*theta)

Wbdotb = (W11*bx + W12*by)*bx + (W21*bx + W22*by)*by

sxx = wbdotb

plt.figure(figsize=(10, 10))
plt.title('Arrows scale with plot width, not view')
Q = plt.quiver(X, Y, U, V, units='width', color='r')
Q2 = plt.quiver(X, Y, bx, by, units='width', color='b')
```
