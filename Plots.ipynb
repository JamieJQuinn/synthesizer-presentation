```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal, fftpack
%matplotlib inline
```

```python
# Number of samplepoints
N = 800
# sample spacing
T = 1.0 / N

x = np.linspace(0.0, N*T, N)
```

```python
y1 = np.sin(5.0 * 2.0*np.pi*x)
y2 = 0.5*np.sin(14.0 * 2.0*np.pi*x)
y = y1 + y2
fig, axes = plt.subplots(3, figsize=(10, 5))
axes[0].plot(x, y1)
axes[1].plot(x, y2, color='r')
axes[1].set_ylim(-1, 1)
axes[2].plot(x, y, color='g')
for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
yf = fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, axes = plt.subplots(1, figsize=(10, 3))

axes.set_xlim(0, 20)
axes.xaxis.set_ticks(range(20))

sns.despine()
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
```
